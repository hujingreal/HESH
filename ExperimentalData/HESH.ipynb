```python
# Motivation：throughput to single DIMM
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 4, 8, 16, 24, 36])
y1 = np.array([1.375401,5.319482,10.277482,18.842023,25.402852,26.217749])
y2 = np.array([0,1.314693667,1.2395,1.070567625,0.820103625,0.067908083])

plt.figure(figsize=(3, 1))
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['font.size'] = 12  

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of threads')
ax1.set_ylabel('Throughput Mops/s',color=color)
ax1.plot(x, y1, color=color, marker='o', zorder=2)
ax1.tick_params(axis='y')
ax1.set_xticks(x)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('The rate of increase',color=color)
ax2.bar(x, y2, color=color, zorder=1, alpha=0.8, width=1)
ax2.tick_params(axis='y')

fig.tight_layout()  
plt.show()
```

```python
#Throughput under YCSB workloads
import matplotlib.pyplot as plt
import numpy as np

with open('dram_pm_ycsb.txt', 'r') as f:
    lines = f.readlines()

data = []
for i in range(0, len(lines), 6):
    data.append([[float(x) for x in line.split()] for line in lines[i:i+6]])

x = [1, 4, 8, 16, 24, 36]

labels = ['HESH', 'Halo', 'Viper', 'SOFT', 'DASH', 'CCEH']

markers = ['o', 'v', '^', '<', '>', 's']

titles = ['(a) I100', '(b) I50 S50', '(c) I5 S95', '(d) S100', '(e) I100', '(f) I50 S50', '(g) SI5 S95', '(h) S100']

fig, axs = plt.subplots(1, 8, figsize=(20, 2.5))

for i, ax in enumerate(axs):
    for j in range(6):
        if i == 0:  
            ax.plot(x, data[i][j], label=labels[j], marker=markers[j])
        else:
            ax.plot(x, data[i][j], marker=markers[j])
    ax.text(0.5, -0.3, titles[i], size=12, ha="center", transform=ax.transAxes)
    ax.grid(axis='y', linestyle='--')
    ax.set_xticks(x)  
    if i == 0:
        ax.set_ylabel('Throughput Mops/s')

fig.legend(ncol=6, loc='upper center', bbox_to_anchor=(0.5,1.08), frameon=False)

plt.tight_layout()
plt.show()
```
